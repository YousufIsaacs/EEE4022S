```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from keras.layers import Dense

columns_to_include = list(range(10))  # This creates a list of integers from 0 to 9
dataset = pd.read_csv('combined_data.csv', delimiter=';', usecols=columns_to_include)

dataset['Timestamp'] = pd.to_datetime(dataset['Timestamp'], format='%Y/%m/%d %H:%M')
```

```python
X_temp = dataset.drop(columns=['Timestamp', dataset.columns[-1]])

y = dataset[dataset.columns[-1]]
```

```python
# Use Random Forest to determine feature importance
rf = RandomForestRegressor()
rf.fit(X_temp, y)
feature_importances = rf.feature_importances_

most_important_feature_indices = np.argsort(feature_importances)[-7:]
most_important_feature_names = X_temp.columns[most_important_feature_indices]

print("Seven Most Important Features:", most_important_feature_names)

# Use only the 7 most important features for training and testing
X_most_important = X_temp.iloc[:, most_important_feature_indices]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_most_important, y, test_size=0.25, random_state=42)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Create and train the neural network
model = tf.keras.models.Sequential([
    Dense(64, activation='relu', input_dim=7),
    Dense(32, activation='relu'),
    Dense(1)
])

model.compile(loss='mse', optimizer='adam', metrics=['RootMeanSquaredError'])
model.summary()
```

```python
# Train the model
model.fit(X_train, y_train, batch_size=32, epochs=500, verbose=2)
```

```python
# Evaluate the model
test_loss, test_rmse = model.evaluate(X_test, y_test)
print("Root Mean Squared Error on Test Data:", test_rmse)
```
